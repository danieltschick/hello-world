###Desafio 06: Renomear as colunas tirando o hífen. 

```python
# reomeando uma unica coluna
dados_renomeados = dados.rename(columns = {'g-0': 'g0', })
dados_renomeados
```

```python
# renomeando todas as colunas com hifen, retirando o hifen
dict_nomes = {}
for i in dados.columns:
  if '-' in i:
    dict_nomes[i] = i.replace('-', '')
dict_nomes

dados_renomeados = dados.rename(columns = dict_nomes)
```

```python
dados_renomeados.columns
```

```python
# Método mais prático: 
dados.columns = dados.columns.str.replace('-', '')
dados.columns
```
